```python
#this chunk of code finds top 30 stations the most travellers
top_stations = pd.DataFrame(df.groupby('STATION')['ENTRIES'].sum())
top_stations = top_stations.sort_values(by = "ENTRIES", ascending=False)
top_stations = top_stations[(top_stations['ENTRIES'] >= top_stations.loc[top_stations.index[29]][0])]
list_of_stations = []
for item in top_stations.index:
    list_of_stations.append(item)
list_of_stations
```

```python
#this chunk of code determines the travellers each day per week of each station
from datetime import datetime
l_items = {}
for item in list_of_stations:
    station = item
    wallstreet = df[df['STATION'] == item]
    wallstreet.columns = wallstreet.columns.str.strip()
    len_index = len(wallstreet)
    wallstreet_dict = {}
    wallstreet_dict_diff = {}
    wallstreet_total_turnstile = {}
    wallstreet = wallstreet.reset_index(drop=True)
#wallstreet dictionary with time and exits as value. Check exercise 7. Add a new item in key which is wallstreet['DATE']!!!!
    for i in range(len_index):
        wallstreet_dict[wallstreet['C/A'][i],wallstreet['UNIT'][i],wallstreet['SCP'][i],wallstreet['STATION'][i],wallstreet['DATE'][i]] = wallstreet_dict.get((wallstreet['C/A'][i],wallstreet['UNIT'][i],wallstreet['SCP'][i],wallstreet['STATION'][i],wallstreet['DATE'][i]),[]) + [(wallstreet['TIME'][i],wallstreet['EXITS'][i])]
#wallstreet_dict_diff. Check exercise 10
    for key,val in wallstreet_dict.items():
        for x in range(len(val)-1):
            wallstreet_dict_diff[key] = wallstreet_dict_diff.get(key, []) + [(val[x+1][0], val[x+1][1]-val[x][1])]
#wallstreet_total_turnstile. Check exercise 12
    for key,val in wallstreet_dict_diff.items():
        total = 0
        for item in val:
            total += item[1]
        wallstreet_total_turnstile[key[4]] = wallstreet_total_turnstile.get(key[4],0) + total
    dt_total = []
    for key,val in wallstreet_total_turnstile.items():
        new_key = pd.to_datetime(key, format='%m/%d/%Y')
        dt_total.append((new_key,val))
    new_dt_total = []
    for item in dt_total:
        if item[1] >= 0:
            new_dt_total.append((item[0].strftime('%a'),item[1]))
    if len(new_dt_total) == 7:
        l_items[station] = l_items.get(station,[]) + new_dt_total
    else:
        l_items[station] = l_items.get(station,[]) + ["ERR"]
print(l_items)
```

```python
#this chunk determines if station is commuter or tourist, with the average number during either weekdays or weekend
stat_dict = {}
for key, val in l_items.items():
    temp = []
    wkdaysum = 0
    wkendsum = 0
    if val != ["ERR"]:
        for i in val:
            if i[0] == "Mon" or i[0] == "Tue" or i[0] == "Wed" or i[0] == "Thu" or i[0] == "Fri":
                wkdaysum += i[1]
            elif i[0] == "Sat" or i[0] == "Sun":
                wkendsum += i[1]
    if (wkdaysum/5) > (wkendsum/2):
        stat_dict[key] = stat_dict.get(key,[]) + ['commuter', wkdaysum/5]
    elif (wkdaysum/5) < (wkendsum/2):
        stat_dict[key] = stat_dict.get(key,[]) + ['tourist', wkendsum/2]
    else:
        stat_dict[key] = stat_dict.get(key,[]) + ['err']
print(stat_dict)
print(len(stat_dict))
```

```python
#sorted top 10 with the most commuters
fin = []
ofin = []
for key, val in stat_dict.items():
    if val[0] == 'commuter':
        fin.append((val[1], key))
fin= sorted(fin, reverse = True)[0:10]
for i in fin:
    ofin.append(i[1])
print(fin)
```

```python
#Andrew's top_hours solution
#the date are
#06/10/2017
#06/11/2017
#06/12/2017
#06/13/2017
#06/14/2017
#06/15/2017
#06/16/2017
import matplotlib.pyplot as plt
def top_hours(station,date):
    ndf = df[(df.STATION == station) & (df.DATE == date)]
    ndf = ndf.reset_index(drop = True) #reset index
    ndf.columns = ndf.columns.str.strip() #column strip
    new_date = pd.to_datetime(date, format='%m/%d/%Y').strftime("%a") #new date in datetime
    day_dict = {}
    for i in range(len(ndf.index)):
        day_dict[(ndf['C/A'][ndf.index[i]],ndf['UNIT'][ndf.index[i]],ndf['SCP'][ndf.index[i]],)] = day_dict.get((ndf['C/A'][ndf.index[i]],ndf['UNIT'][ndf.index[i]],ndf['SCP'][ndf.index[i]]),[]) + [(ndf['TIME'][ndf.index[i]],ndf['EXITS'][ndf.index[i]])]
    day_dict_diff = {}
    for key,val in day_dict.items():
        for x in range(len(val)-1):
            day_dict_diff[key] = day_dict_diff.get(key, []) + [(val[x+1][0], val[x+1][1]-val[x][1])]
    new_list = {}
    for key,val in day_dict_diff.items():
        for item in val:
            new_list[item[0]] = new_list.get(item[0],0) + item[1]
    return new_list
days_of_the_week = ['06/10/2017','06/11/2017','06/12/2017','06/13/2017','06/14/2017','06/15/2017','06/16/2017']

def top_hours_2(station):
    temp_dict = {}
    for day in days_of_the_week:
        for key,val in top_hours(station,day).items():
            temp_dict[key] = temp_dict.get(key,[]) + [val]
    temp_dict_2 = {}
    for key,val in temp_dict.items():
        a = 0
        for item in val:
            a += item
        average = a/len(val)
        temp_dict_2[key] = temp_dict_2.get(key,0) + average
    return temp_dict_2
```
